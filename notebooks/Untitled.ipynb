```python
def create_plot(model,sounds,sound_compiled,fakes,metrics_names,refresh_rate,save_name,create_anim = True):
    
    # sound_compiled = sounds.reshape(-1)
    nb_samples = sounds.shape[0]
    len_total_sec = sound_compiled.shape[0]/model.operator.sr
    len_sample_sec = len_total_sec/nb_samples
    len_sample_ech = int(len_sample_sec*model.operator.sr)
    print(len_sample_ech)

    metrics = compute_metrics(sound_compiled,metrics_names)
    
    fig = plt.figure(figsize=(18,10))
    spec = fig.add_gridspec(6,1,height_ratios = [0.75, 0.25, 0.15,0.15,0.15,0.15])
    ax_quad = fig.add_subplot(spec[0, :])
    ax_signal = fig.add_subplot(spec[1, :])
    ax0 = fig.add_subplot(spec[2, :])
    ax1 = fig.add_subplot(spec[3, :])
    ax2 = fig.add_subplot(spec[4, :])
    ax3 = fig.add_subplot(spec[5, :])
    ax_metrics = [ax0,ax1,ax2,ax3]
    
    
    #SPECTROGRAM
    times_spec = np.linspace(0, 2 , fakes[0].shape[1])
    freqs_spec = librosa.mel_frequencies(
        n_mels=fakes[0].shape[0], fmin=0.0, fmax=16000 / 2, htk=True
    ) 
    def add_colorbar(fig, ax, im):
        divider = make_axes_locatable(ax)
        cax = divider.append_axes("right", size="5%", pad=0.05)
        fig.colorbar(im, cax=cax, orientation="vertical")
    X, Y = np.meshgrid(times_spec, freqs_spec)
    im0 = ax_quad.pcolormesh(X, Y, fakes[0], cmap="magma",vmin=-1,vmax=1)
    add_colorbar(fig, ax_quad, im0)
    ax_quad.set_ylabel('Frequency')
    ax_quad.set_title('Spectrogram')
    ax_quad.set_xticks([])
    
    #WAVEFORM
    
    # SIGNAL
    total_times = np.linspace(0,len_total_sec,sound_compiled.shape[0])
    ax_signal.set_xlim(0,total_times[-1])
    color = ['whitesmoke','grey']
    ax_signal.set_xticks([])
    ax_signal.set_ylabel('Waveform',weight='bold')
    
    ax_signal.fill_between(total_times,sound_compiled)
    
    for j in range(nb_samples):
        ax_signal.axvspan(len_sample_sec*j,len_sample_sec*(j+1),alpha=0.075,color=color[j%2])
        
    
        
    #METRICS  
    for n,ax in enumerate(ax_metrics):
        name = metrics_names[n]
        ax.set_ylabel(name,weight='bold')
        ax.set_xlim(0,len_total_sec)
        
        if n!=len(ax_metrics)-1:
            ax.set_xticks([])
        else:
            ax.set_xlabel('Time (s)')
            
        metric = metrics[name][0]
        
        metric_times = np.linspace(0,len_total_sec,metric.shape[0])
        metric_step = metric.shape[0]/nb_samples

        ax.plot(metric_times,metric,color='navy',alpha=0.7,linewidth=0.7)
        
        for j in range(nb_samples):
            istart = int(j*metric_step)
            iend = int((j+1)*metric_step)
            mean_value = np.mean(metric[istart:iend])
            
            ax.axvspan(len_sample_sec*j,len_sample_sec*(j+1),alpha=0.075,color=color[j%2])
            ax.hlines(mean_value,len_sample_sec*j,len_sample_sec*(j+1),color='k',linewidth=0.85)
            
            
    vl = ax_signal.axvline(0, ls='-', color='r',alpha=0.5,linewidth=0.9) 
    vl1 = ax0.axvline(0, ls='-', color='r',alpha=0.5,linewidth=0.9)
    vl2 = ax1.axvline(0, ls='-', color='r',alpha=0.5,linewidth=0.9)
    vl3 = ax2.axvline(0, ls='-', color='r',alpha=0.5,linewidth=0.9)
    vl0 = ax3.axvline(0, ls='-', color='r',alpha=0.5,linewidth=0.9)    
    
    fig.align_ylabels([ax_quad,ax_signal,ax0,ax1,ax2,ax3])
    
   
    if create_anim == True:
        time_step_ech = int(len_sample_ech/refresh_rate)
        time_step_sec = time_step_ech/model.operator.sr
        nb_frames = refresh_rate * nb_samples

        def animate(i):
            step = time_step_ech*i
            j = (step+200)//len_sample_ech
            time = step/model.operator.sr
            if ((step+200)%len_sample_ech) <= len_sample_ech/4/refresh_rate:
                ax_quad.set_title('Sample :' + str(j))
                im0.set_array(fakes[j][:-1, :-1])
                j+=1

            vl.set_xdata([time,time])

            # for line in vlines_metrics:
            vl0.set_xdata([time,time])
            vl1.set_xdata([time,time])
            vl2.set_xdata([time,time])
            vl3.set_xdata([time,time])

            return im0,vl,vl1,vl2,vl3,vl0

        def init():
            pass
        
        
        frame_interval = time_step_sec
        fps = 1/frame_interval
        anim = animation.FuncAnimation(fig,animate,frames=nb_frames,interval=frame_interval*1000,blit=True,repeat=False,init_func=init,cache_frame_data=False)
        plt.close(fig)

        audio_file = os.path.join(save_path,"audio.wav")
        video_file = os.path.join(save_path,"video.mp4")
        final_file = os.path.join(save_path,save_name)
        
        plt.rcParams['animation.ffmpeg_path'] = os.path.join(ffmpeg_path,'ffmpeg.exe')
        writer = animation.FFMpegWriter(fps=fps,bitrate=250)#, codec='libx264') # Or 
        anim.save(video_file, writer=writer)

        
        soundfile.write(audio_file, sound_compiled,model.operator.sr)
        input_video = ffmpeg.input(video_file)
        input_audio = ffmpeg.input(audio_file)
        
        os.chdir(ffmpeg_path)

        ffmpeg.concat(input_video, input_audio, v=1, a=1).output(final_file).run()

```
